```python
ab = df["Summary-摘要"]
ab.head()
```

```python
file = open("ab.txt","w",encoding="utf-8")
for word in ab:
    file.write(word)
file.close()
```

```python
import jieba
file1 = open("ab.txt",encoding="utf-8")
text = file1.read()
jieba.add_word("深度学习")
jieba.add_word("循环神经网络")
jieba.add_word("卷积神经网络")
jieba.add_word("循环神经网络")
fenci = jieba.lcut(text.strip())
new = list(fenci)

print(new)

# file = open("fenci.txt","w",encoding="utf-8")

# file.write(str(fenci))
# file.close()
```

```python
#去停用词
after_stopwords = []
stopwords_file = open("stopwords.txt","r",encoding="utf-8")
stopwords = [line.strip() for line in stopwords_file.readlines()]
# print(stopwords)
for i in new:
    i = i.strip()
    if i not in stopwords:
        if i != "\t":
            after_stopwords.append(i)
#             after_stopwords += " "
print("去除停用词后：",after_stopwords)
```

```python
# 统计词频

from collections import Counter
count = Counter(after_stopwords)
print(count.most_common(50))
```
